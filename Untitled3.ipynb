```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation # New import for animation
import time
import heapq
import random
from collections import defaultdict
import pandas as pd
import traceback # For detailed error tracing
import os # For creating directories for animation frames

class MSTVisualizer:
    """Enhanced MST visualizer with all 5 algorithms and improved visualization"""

    # New thresholds for static plot generation
    STATIC_PLOT_NODE_THRESHOLD = 4000
    STATIC_PLOT_EDGE_THRESHOLD = 13000

    def __init__(self, graph_file=None, initial_graph=None):
        self.graph_file = graph_file
        self.G = None # The NetworkX graph
        self.node_to_idx = {} # Maps node names (from graph file) to integer indices (for DSU)
        self.idx_to_node = {} # Maps integer indices back to node names
        self.pos = None # Stores node positions for consistent plotting/animation

        # Data lists to store historical data for analytical plots generated *after* algorithm execution
        self.cost_points = []
        self.edges_count_points = []
        self.components_count_points = []

        if graph_file:
            self.load_graph(graph_file)
        elif initial_graph is not None: # Corrected `!==` to `is not None`
            self._set_graph(initial_graph)

    def _set_graph(self, graph):
        """Internal method to set an already constructed NetworkX graph."""
        self.G = graph
        # Ensure all nodes are consistently converted to strings *before* creating node_to_idx
        self.G = nx.relabel_nodes(self.G, {node: str(node) for node in self.G.nodes()})

        # Ensure all edges have weights as floats
        for u, v, data in self.G.edges(data=True):
            if 'weight' not in data:
                data['weight'] = random.uniform(0.1, 10.0)
            else:
                try:
                    data['weight'] = float(data['weight'])
                except (ValueError, TypeError):
                    data['weight'] = random.uniform(0.1, 10.0)

        self.node_to_idx = {node: i for i, node in enumerate(self.G.nodes())}
        self.idx_to_node = {i: node for node, i in self.node_to_idx.items()}
        print(f"Graph set with {len(self.G.nodes())} nodes and {len(self.G.edges())} edges.")
        if len(self.G.nodes()) == 0:
            raise ValueError("Provided an empty graph. Please provide a graph with at least one node.")

        # Calculate and store positions once for consistent layouts across all plots/animations
        self.pos = nx.spring_layout(self.G, k=0.3, iterations=100, seed=42)

    def load_graph(self, file_path):
        """
        Loads a graph from a specified file. Supports CSV and edgelist formats.
        Assigns random weights if the graph is unweighted.
        Handles disconnected graphs by taking the largest connected component.
        """
        print(f"\nLoading graph from: {file_path}")

        try:
            if file_path.endswith('.csv'):
                df = pd.read_csv(file_path)
                # Check if enough columns for weighted edgelist (source, target, weight)
                if len(df.columns) >= 3:
                    temp_G = nx.from_pandas_edgelist(df,
                                                     source=df.columns[0],
                                                     target=df.columns[1],
                                                     edge_attr=df.columns[2])
                    print("Loaded as weighted CSV.")
                else:
                    # If only two columns, assume unweighted and assign random weights
                    temp_G = nx.from_pandas_edgelist(df,
                                                     source=df.columns[0],
                                                     target=df.columns[1])
                    print("Loaded as unweighted CSV - assigning random weights.")
                    for u, v in temp_G.edges():
                        temp_G[u][v]['weight'] = random.uniform(0.1, 10.0)
            else: # If not CSV, assume edgelist format
                try:
                    temp_G = nx.read_weighted_edgelist(file_path)
                    print("Loaded as weighted edgelist.")
                except:
                    # If read_weighted_edgelist fails, try as unweighted and assign random weights
                    temp_G = nx.read_edgelist(file_path)
                    print("Loaded as unweighted edgelist - assigning random weights.")
                    for u, v in temp_G.edges():
                        temp_G[u][v]['weight'] = random.uniform(0.1, 10.0)

            # Handle disconnected graphs: Work only on the largest connected component
            if not nx.is_connected(temp_G):
                print("Graph is not connected - using largest connected component.")
                largest_cc = max(nx.connected_components(temp_G), key=len)
                temp_G = temp_G.subgraph(largest_cc).copy() # Use .copy() to ensure it's a new graph

            self._set_graph(temp_G) # Use the new internal method to set the graph and mappings

        except Exception as e:
            print(f"Failed to load graph: {e}")
            self.G = nx.Graph() # Initialize an empty graph to prevent further errors
            raise # Re-raise the exception to be caught by the main block for specific error messages

    def _plot_final_mst(self, mst_edges, title, cost):
        """
        Generates and displays a static plot of the final MST.
        """
        plt.close('all') # Close any previous figures
        fig, ax = plt.subplots(1, 1, figsize=(10, 8), layout='constrained')

        fig.patch.set_facecolor('white')
        ax.set_facecolor('#f9f9f9')
        ax.set_title(f'Final MST - {title}\nCost: {cost:.4f}', fontsize=16, fontweight='bold', color='#333333')

        # Use the class's pre-calculated positions for consistency
        pos = self.pos

        # Draw all graph edges as a light gray background
        nx.draw_networkx_edges(self.G, pos, ax=ax, edge_color='lightgray', alpha=0.4, width=1.0)

        # Draw all nodes (default color)
        nx.draw_networkx_nodes(self.G, pos, ax=ax, node_size=50,
                               node_color='skyblue', alpha=0.9, linewidths=1.0, edgecolors='black')

        # Draw node labels (made smaller as requested)
        nx.draw_networkx_labels(self.G, pos, ax=ax, font_size=3, font_weight='bold', font_color='darkslategray') # Changed font_size to 3

        # Draw MST edges in black (on top) - requested change
        final_mst_edgelist = [(str(u), str(v)) for u, v, _ in mst_edges]
        nx.draw_networkx_edges(self.G, pos, edgelist=final_mst_edgelist, ax=ax,
                               edge_color='black', width=4.0, alpha=0.9, style='solid') # Changed to black and wider

        ax.set_axis_on() # Ensure axes are visible for static plot
        ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True, colors='black')
        ax.grid(True, color='#cccccc', linestyle='--', alpha=0.7)

        plt.tight_layout()
        plt.show(block=False) # Show plot without blocking script execution

    def kruskal_mst(self):
        """
        Implements Kruskal's algorithm to find the Minimum Spanning Tree.
        Uses a Disjoint Set Union (DSU) data structure.
        """
        start_time = time.time() # Start timing for the algorithm's logic

        edges = [(self.G[u][v]['weight'], u, v) for u, v in self.G.edges()]
        edges.sort() # Sort all edges by weight in ascending order

        n = len(self.G.nodes())
        dsu = DisjointSetUnion(n) # Initialize DSU with all nodes in their own sets

        mst_edges = [] # List to store edges included in the MST
        mst_cost = 0 # Total cost of the MST

        # Data lists for tracking state evolution for analytical plots
        cost_evolution = [0]
        edges_count_evolution = [0]
        components_count_evolution = [n] # Initially, each node is a component

        step_counter = 0
        # Core algorithm loop starts here
        for weight, u, v in edges:
            u_str = str(u)
            v_str = str(v)
            if dsu.union(self.node_to_idx[u_str], self.node_to_idx[v_str]):
                mst_edges.append((u, v, weight)) # Add edge to MST
                mst_cost += float(weight) # Explicitly cast weight to float before arithmetic

                step_counter += 1
                # Record current state for evolution plots
                cost_evolution.append(mst_cost)
                edges_count_evolution.append(len(mst_edges))
                components_count_evolution.append(dsu.num_components)

                # MST is complete when it has N-1 edges (for a connected graph with N nodes)
                if len(mst_edges) == n - 1:
                    break

        # Core algorithm loop ends here
        execution_time = time.time() - start_time # End timing here

        return mst_edges, mst_cost, execution_time, cost_evolution, edges_count_evolution, components_count_evolution

    def prim_mst(self):
        """
        Implements Prim's algorithm to find the Minimum Spanning Tree.
        Uses a min-priority queue (heap).
        """
        start_time = time.time() # Start timing for the algorithm's logic

        if not self.G.nodes(): # Handle empty graph
            return [], 0, 0, [0], [0], [0] # Return empty evolution data too

        start_node = next(iter(self.G.nodes())) # Pick an arbitrary start node (first node in graph)

        # Min-priority queue: stores (edge_weight, from_node, to_node)
        # It holds edges that connect a visited node to an unvisited node.
        pq = []

        visited = {start_node} # Set of nodes already included in the MST
        mst_edges = [] # List to store edges included in the MST
        mst_cost = 0 # Total cost of the MST

        # Data lists for tracking state evolution for analytical plots
        cost_evolution = [0]
        edges_count_evolution = [0]
        components_count_evolution = [len(self.G.nodes())] # Initially, N components

        step_counter = 0
        # Add all edges connected to the start_node to the priority queue
        for neighbor in self.G.neighbors(start_node):
            weight = self.G[start_node][neighbor]['weight']
            heapq.heappush(pq, (weight, start_node, neighbor))

        # Core algorithm loop starts here
        # Continue as long as there are edges in PQ and we haven't included all nodes
        while pq and len(visited) < len(self.G.nodes()):
            weight, u, v = heapq.heappop(pq) # Get the cheapest edge

            if v in visited: # If 'v' is already in the MST, this edge forms a cycle, so skip
                continue

            # Add 'v' to visited, and the edge (u,v) to MST
            visited.add(v)
            mst_edges.append((u, v, weight))
            mst_cost += float(weight) # Explicitly cast weight to float before arithmetic

            step_counter += 1
            # Record current state for evolution plots
            cost_evolution.append(mst_cost)
            edges_count_evolution.append(len(mst_edges))
            components_count_evolution.append(len(self.G.nodes()) - len(mst_edges)) # Prim's reduces components by 1 for each edge added

            # For the newly added node 'v', add all its unvisited neighbors' edges to the PQ
            for neighbor_of_v in self.G.neighbors(v):
                if neighbor_of_v not in visited:
                    edge_weight = self.G[v][neighbor_of_v]['weight']
                    heapq.heappush(pq, (edge_weight, v, neighbor_of_v))

        # Core algorithm loop ends here
        execution_time = time.time() - start_time # End timing here

        return mst_edges, mst_cost, execution_time, cost_evolution, edges_count_evolution, components_count_evolution

    def boruvka_mst(self):
        """
        Implements Borůvka's algorithm to find the Minimum Spanning Tree.
        Iteratively adds the cheapest edge leaving each connected component.
        """
        start_time = time.time() # Start timing for the algorithm's logic

        n = len(self.G.nodes())
        if n == 0:
            return [], 0, 0, [0], [0], [0] # Return empty evolution data too

        dsu = DisjointSetUnion(n) # DSU to keep track of connected components
        mst_edges = []
        mst_cost = 0

        # Data lists for tracking state evolution for analytical plots
        cost_evolution = [0]
        edges_count_evolution = [0]
        components_count_evolution = [n] # Start with N components

        step_counter = 0
        # Core algorithm loop starts here
        # Continue phases until only one component remains (the MST is formed)
        while dsu.num_components > 1:
            # Stores the minimum weight edge found for each component in the current phase
            # Format: {component_root_idx: (weight, u_node, v_node)}
            component_min_edge = {}

            # Iterate through all edges in the graph to find the cheapest outgoing edge for each component
            for u_node, v_node in self.G.edges():
                weight = self.G[u_node][v_node]['weight']
                # Ensure u_node and v_node are passed as strings to self.node_to_idx
                u_str = str(u_node)
                v_str = str(v_node)
                comp_u_root = dsu.find(self.node_to_idx[u_str]) # Find root of component for u
                comp_v_root = dsu.find(self.node_to_idx[v_str]) # Find root of component for v

                if comp_u_root != comp_v_root: # If edge connects two different components
                    # Update minimum edge for component of u
                    if comp_u_root not in component_min_edge or float(weight) < component_min_edge[comp_u_root][0]: # Compare as float
                        component_min_edge[comp_u_root] = (float(weight), u_node, v_node) # Store as float
                    # Update minimum edge for component of v
                    if comp_v_root not in component_min_edge or float(weight) < component_min_edge[comp_v_root][0]: # Compare as float
                        component_min_edge[comp_v_root] = (float(weight), u_node, v_node) # Store as float

            if not component_min_edge: # If no edges connect different components, the graph is effectively disconnected
                break # or the MST is already complete if num_components == 1

            # Add the cheapest edges (one per component) to the MST.
            # It's important to process these edges in a single phase.
            edges_added_in_phase = False
            for comp_root, (weight, u_node, v_node) in component_min_edge.items():
                # Ensure u_node and v_node are passed as strings to self.node_to_idx
                u_str = str(u_node)
                v_str = str(v_node)
                # Attempt to union the components. A union only occurs if they are still distinct.
                if dsu.union(self.node_to_idx[u_str], self.node_to_idx[v_str]):
                    mst_edges.append((u_node, v_node, weight))
                    mst_cost += float(weight) # Explicitly cast weight to float before arithmetic
                    edges_added_in_phase = True # Flag that progress was made
                    step_counter += 1

                    # Record current state for evolution plots
                    cost_evolution.append(mst_cost)
                    edges_count_evolution.append(len(mst_edges))
                    components_count_evolution.append(dsu.num_components)

            # If no edges were added in a phase but there are still multiple components,
            # it means the remaining components are disconnected from each other.
            if not edges_added_in_phase and dsu.num_components > 1:
                print("Warning: Boruvka's algorithm made no progress in a phase but components remain. Graph may be disconnected or all remaining edges are cycles.")
                break
        # Core algorithm loop ends here
        execution_time = time.time() - start_time # End timing here

        return mst_edges, mst_cost, execution_time, cost_evolution, edges_count_evolution, components_count_evolution

    def reverse_delete_mst(self):
        """
        Implements the Reverse-Delete algorithm to find the Minimum Spanning Tree.
        Starts with all edges and iteratively removes the heaviest edges that do not disconnect the graph.
        Optimized by using nx.has_path for connectivity checks.
        """
        start_time = time.time() # Start timing for the algorithm's logic

        # Get all edges with their weights and sort them in descending order of weight
        all_graph_edges_sorted_desc = [(self.G[u][v]['weight'], u, v) for u, v in self.G.edges()]
        all_graph_edges_sorted_desc.sort(key=lambda x: x[0], reverse=True)

        # Start with all edges conceptually part of the MST and their total cost
        mst_edges = [(u, v, w) for w, u, v in all_graph_edges_sorted_desc]
        mst_cost = sum(float(w) for w, _, _ in all_graph_edges_sorted_desc) # Ensure sum is done with floats

        # Create a mutable copy of the graph to remove/add edges for connectivity checks
        temp_graph = self.G.copy()

        # Data lists for tracking state evolution for analytical plots
        cost_evolution = [mst_cost]
        edges_count_evolution = [len(mst_edges)]
        components_count_evolution = [1] # Graph starts as connected, should remain 1

        step_counter = 0
        # Core algorithm loop starts here
        for weight, u, v in all_graph_edges_sorted_desc:
            # Only process if the edge still exists in temp_graph
            if not temp_graph.has_edge(u, v):
                continue

            # Temporarily remove the current edge from the working graph
            temp_graph.remove_edge(u, v)

            # Check if u and v are still connected *without* this edge.
            # nx.has_path(graph, source, target) is much more efficient than nx.is_connected(graph)
            # because it stops as soon as a path is found.
            # Ensure u and v are passed as strings to nx.has_path
            if nx.has_path(temp_graph, str(u), str(v)):
                # If they are still connected, this edge is redundant (it creates a cycle),
                # so it can be safely removed from the final MST.
                mst_edges.remove((u, v, weight)) # Remove from our conceptual MST list
                mst_cost -= float(weight) # Explicitly cast weight to float before arithmetic

                step_counter += 1
                # Record current state for evolution plots
                cost_evolution.append(mst_cost)
                edges_count_evolution.append(len(mst_edges))
                components_count_evolution.append(1) # Should remain 1
            else:
                # If removing it disconnects the graph (u and v are no longer connected),
                # then this edge is essential for connectivity and must be part of the MST.
                # So, add it back to the temporary graph (it was "retained").
                temp_graph.add_edge(u, v, weight=weight)
                # No change to mst_edges or mst_cost, as it was never removed from final consideration
        # Core algorithm loop ends here
        execution_time = time.time() - start_time # End timing here

        return mst_edges, mst_cost, execution_time, cost_evolution, edges_count_evolution, components_count_evolution

    def karger_mst(self):
        """
        Implements a "Randomized Kruskal-like" algorithm for finding MST.
        This adapts Karger's idea of randomness for the MST problem.
        It runs multiple trials, sorting edges by weight then shuffling edges of equal weight,
        and returns the best (lowest cost) MST found across trials.
        This is not the exact Karger's min-cut algorithm, but a common interpretation for MST.
        """
        start_time = time.time() # Start timing for the algorithm's logic

        if len(self.G.nodes()) < 2: # Handle trivial graph
            return [], 0, 0, [0], [0], [0] # Return empty evolution data too

        # Ensure all node names are converted to strings consistently and weights are floats.
        all_graph_edges = []
        for u, v in self.G.edges():
            u_str = str(u) # Ensure u is a string for adding to all_graph_edges
            v_str = str(v) # Ensure v is a string for adding to all_graph_edges
            weight = float(self.G[u][v]['weight'])
            all_graph_edges.append((u_str, v_str, weight))

        n = len(self.G.nodes())

        # Variables to store the best MST found across all trials
        best_mst_edges = []
        best_mst_cost = float('inf')
        best_cost_evolution = []
        best_edges_count_evolution = []
        best_components_count_evolution = []

        num_trials = 5 # Number of random trials to run to increase probability of finding true MST

        print(f"Running Karger's (Randomized Kruskal) with {num_trials} trials to find the best MST...")

        # Core algorithm loop starts here (encompassing all trials)
        for trial in range(num_trials):
            dsu = DisjointSetUnion(n)
            current_mst_edges = []
            current_mst_cost = 0

            trial_cost_evolution = [0]
            trial_edges_count_evolution = [0]
            trial_components_count_evolution = [n]

            # --- Randomized Edge Selection Strategy ---
            # 1. Sort all edges by weight.
            # 2. Group edges by weight, then randomly shuffle within each weight group.
            # This ensures lighter edges are generally considered first, but their order
            # (when weights are equal) is randomized.

            # Sort all edges first
            edges_for_trial = sorted(all_graph_edges, key=lambda x: x[2])

            # Group by weight
            weight_groups = defaultdict(list)
            for u, v, w in edges_for_trial:
                weight_groups[w].append((u, v, w))

            shuffled_edges = []
            # Iterate through weights in sorted order, then shuffle edges within each weight group
            for weight_key in sorted(weight_groups.keys()):
                random.shuffle(weight_groups[weight_key]) # Shuffle edges with the same weight
                shuffled_edges.extend(weight_groups[weight_key]) # Add to the final shuffled list

            step_counter_trial = 0

            for u, v, weight in shuffled_edges:
                # Explicitly convert u and v to string before using as dictionary keys
                # This is the crucial fix for the KeyError '1.0' vs '1' type mismatch
                u_idx = self.node_to_idx[str(u)]
                v_idx = self.node_to_idx[str(v)]

                # If union is successful (edge connects two different components)
                if dsu.union(u_idx, v_idx):
                    current_mst_edges.append((u, v, weight))
                    current_mst_cost += float(weight)

                    step_counter_trial += 1
                    # Record current state for this trial's evolution
                    trial_cost_evolution.append(current_mst_cost)
                    trial_edges_count_evolution.append(len(current_mst_edges))
                    trial_components_count_evolution.append(dsu.num_components)

                if len(current_mst_edges) == n - 1: # MST is complete for this trial
                    break

            # After each trial, check if a complete MST was found and if it's better than the best so far
            # Only update best_animation_frames_data if a new best MST is found
            if len(current_mst_edges) == n - 1 and current_mst_cost < best_mst_cost:
                best_mst_cost = current_mst_cost
                best_mst_edges = current_mst_edges
                best_cost_evolution = trial_cost_evolution
                best_edges_count_evolution = trial_edges_count_evolution
                best_components_count_evolution = trial_components_count_evolution
            # No explicit warning for disconnected trials as it is expected in randomized algos
        # Core algorithm loop ends here
        execution_time = time.time() - start_time # End timing here

        if not best_mst_edges and n > 1:
            print("Warning: Karger's (Randomized Kruskal) algorithm failed to find a complete MST after all trials.")
            return [], 0, 0, [0], [0], [0] # Return empty evolution data too
        elif not best_mst_edges and n <= 1: # Trivial case (empty or single node graph)
            return [], 0, 0, [0], [0], [0] # Return empty evolution data too

        return best_mst_edges, best_mst_cost, execution_time, best_cost_evolution, best_edges_count_evolution, best_components_count_evolution

    def animate_mst(self, mst_graph_nx, title="MST Animation", save_path=None):
        """
        Generates an animation of the MST growing.
        The animation adds edges in batches based on a 'step' threshold.
        """
        fig, ax = plt.subplots(figsize=(10, 7))
        # Use the pre-calculated positions from the class for consistency
        pos = self.pos

        # Ensure the mst_graph_nx has nodes from the original graph, just with MST edges
        # This is crucial for drawing all nodes in the animation, even if they're not yet connected by an MST edge.
        # It's also important to use original_graph.nodes() for consistency.
        nx.draw_networkx_nodes(self.G, pos, ax=ax, node_size=50, node_color='skyblue',
                               linewidths=1.0, edgecolors='black') # Draw all original nodes
        nx.draw_networkx_labels(self.G, pos, ax=ax, font_size=3, font_weight='bold',
                                font_color='darkslategray') # Draw all original labels (smaller)

        # Draw all original graph edges as a light background
        nx.draw_networkx_edges(self.G, pos, ax=ax, edge_color='lightgray', alpha=0.4, width=1.0)


        # Sort MST edges by weight for consistent animation growth
        edges_sorted = sorted(list(mst_graph_nx.edges(data=True)), key=lambda x: x[2]['weight'])

        frames = []
        temp_graph = nx.Graph()
        # Add all nodes from the original graph to temp_graph to ensure they are always present
        temp_graph.add_nodes_from(self.G.nodes())

        # 'connected_nodes' from your original snippet is somewhat misleading for the animation logic here.
        # Instead, we just add edges and check the count of edges in the temp_graph.
        # The animation is meant to show edges accumulating.

        # Changed 'step' to be based on number of edges for more even animation steps
        # If total edges are 914 and step is 100, frames will be at roughly 100, 200, ..., 900 edges.
        animation_step_interval = 10
        next_edge_count_threshold = animation_step_interval

        for u, v, data in edges_sorted: # Iterate through sorted MST edges
            temp_graph.add_edge(u, v, **data) # Add edge to the temporary graph

            # Capture frame when the number of edges added reaches the threshold
            if len(temp_graph.edges()) >= next_edge_count_threshold:
                frames.append(list(temp_graph.edges(data=True))) # Add current state of temp_graph edges
                next_edge_count_threshold += animation_step_interval

        # Ensure the final state is always captured, even if it doesn't hit a threshold exactly
        if not frames or list(temp_graph.edges(data=True)) not in frames:
             frames.append(list(temp_graph.edges(data=True)))

        def update(frame_index):
            ax.clear()
            ax.set_title(title + f"\nEdges: {len(frames[frame_index])} / {len(edges_sorted)}", fontsize=14, color='#555555')

            # Re-draw static background elements for each frame
            nx.draw_networkx_nodes(self.G, pos, ax=ax, node_size=50, node_color='skyblue',
                                   linewidths=1.0, edgecolors='black')
            nx.draw_networkx_labels(self.G, pos, ax=ax, font_size=3, font_weight='bold',
                                    font_color='darkslategray')
            nx.draw_networkx_edges(self.G, pos, ax=ax, edge_color='lightgray', alpha=0.4, width=1.0)

            # Draw the current set of MST edges for this frame
            frame_edges = frames[frame_index]
            nx.draw_networkx_edges(self.G, pos, ax=ax, edgelist=frame_edges,
                                   edge_color='green', width=2, alpha=0.9) # Green and width 2 as per your snippet

            # Adjust plot limits for consistency
            ax.set_xlim(ax.get_xlim())
            ax.set_ylim(ax.get_ylim())

            return ax, # Return a tuple of artists (required by FuncAnimation)

        # Adjusted interval for smoother animation
        ani = FuncAnimation(fig, update, frames=len(frames), interval=150, blit=False, repeat=False)

        if save_path:
            print(f"Saving animation to {save_path}...")
            try:
                ani.save(save_path, writer='pillow', fps=10) # Added fps for better GIF quality
                print(f"Animation saved to {save_path}")
            except Exception as e:
                print(f"Error saving animation: {e}")
                print(f"Please ensure 'Pillow' is installed (pip install Pillow) and FFmpeg is available if using 'ffmpeg' writer.")

        plt.close(fig) # Close the figure to prevent it from showing up as a separate window after saving
        return ani

    def save_mst_animation(self, mst_graphs_dict):
        """
        Iterates through generated MST graphs and saves their animations as GIFs.
        Only processes graphs within the animation size thresholds.
        """
        # Define thresholds for animation (using STATIC_PLOT_ thresholds as a proxy for animation capability)
        # This means animation is only attempted for graphs small enough for static plots.
        if len(self.G.nodes()) > self.STATIC_PLOT_NODE_THRESHOLD or \
           len(self.G.edges()) > self.STATIC_PLOT_EDGE_THRESHOLD:
            print(f"\n--- Skipping animation saving for current graph ---")
            print(f"  Graph has {len(self.G.nodes())} nodes and {len(self.G.edges())} edges, which is too large for animation.")
            print(f"----------------------------------------------------\n")
            return

        for name, mst_graph_nx in mst_graphs_dict.items():
            # The original request was to skip Karger, so keeping that here.
            # If you want Karger animated too, remove this condition.
            if name != "Karger'":
                # Create a filename based on the algorithm name
                filename = name.lower().replace(' ', '_').replace("'", "") + "_mst_animation.gif"

                # Check if the MST graph is empty before attempting animation
                if mst_graph_nx.number_of_edges() == 0 and len(self.G.nodes()) > 1:
                    print(f"Skipping animation for {name}: MST graph is empty or trivial for this algorithm.")
                    continue

                print(f"Saving animation for {name}...")
                # Pass the NetworkX graph object (mst_graph_nx) and the class's pre-calculated positions
                self.animate_mst(mst_graph_nx, title=name + " MST", save_path=filename)
            else:
                print(f"Skipping animation for {name} as per previous request.")


    def compare_all_algorithms(self):
        """
        Compares all 5 implemented MST algorithms, and then generates comprehensive analytical plots for comparison.
        Now also includes optional animation generation for smaller graphs.
        """
        if self.G is None or len(self.G.nodes()) == 0:
            print("No graph loaded or graph is empty. Please load a valid graph first.")
            return {}

        # Define the algorithms to compare.
        algorithms = [
            ("Kruskal's Algorithm", self.kruskal_mst),
            ("Prim's Algorithm", self.prim_mst),
            ("Borůvka's Algorithm", self.boruvka_mst),
            ("Reverse-Delete Algorithm", self.reverse_delete_mst),
            ("Karger's Algorithm", self.karger_mst) # Labelled as Karger's, but implemented as Randomized Kruskal
        ]

        results = {} # Stores final cost, time, number of edges, and steps for comparison table
        all_alg_states = {} # Stores all evolution data (cost, edges, components) for comparative plots
        mst_graphs_for_animation = {} # Stores NetworkX MST graphs for animation generation

        print("\n=== Comparing All MST Algorithms ===")
        print("="*70)

        self.plot_degree_distribution() # This is a static plot, always shown

        for name, algorithm_func in algorithms: # Unpack the new tuple
            # Skip Reverse-Delete for large graphs
            if name == "Reverse-Delete Algorithm" and \
               (len(self.G.nodes()) > self.STATIC_PLOT_NODE_THRESHOLD or len(self.G.edges()) > self.STATIC_PLOT_EDGE_THRESHOLD):
                print(f"\n--- Skipping {name} for large graph ({len(self.G.nodes())} nodes, {len(self.G.edges())} edges) ---")
                print(f"  {name} can be exceptionally slow on large graphs due to repeated connectivity checks.")
                print(f"  Proceeding with the next algorithm.")
                print(f"----------------------------------------------------------------------------------\n")
                results[name] = {'skipped': True} # Mark as skipped
                continue

            print(f"\nRunning {name}...")

            try:
                # Call algorithm function (now returns just evolution data)
                edges, cost, exec_time, cost_evo, edges_evo, comp_evo = algorithm_func()

                # Debug print to see if edges are returned for Reverse-Delete
                if name == "Reverse-Delete Algorithm":
                    print(f"  Reverse-Delete returned {len(edges)} edges with cost {cost:.4f}.")

                # Verify the results against NetworkX's MST
                self._verify_mst(edges, cost)

                # Store results for the comparison table
                results[name] = {
                    'edges': edges,
                    'cost': cost,
                    'time': exec_time,
                    'steps': len(cost_evo)-1 # Number of distinct steps where state changed
                }
                # Store all evolution data for later plotting on single comparative graphs
                all_alg_states[name] = {
                    'cost_evolution': cost_evo,
                    'edges_evolution': edges_evo,
                    'components_evolution': comp_evo
                }

                # Create a NetworkX graph from the found MST edges for animation (if applicable)
                if edges: # Only create if MST was found
                    mst_graph_nx = nx.Graph()
                    mst_graph_nx.add_nodes_from(self.G.nodes()) # Add all nodes from original graph
                    mst_graph_nx.add_weighted_edges_from(edges)
                    mst_graphs_for_animation[name] = mst_graph_nx

                print(f"  Final MST Cost: {cost:.4f}")
                print(f"  Execution Time (Core Algorithm): {exec_time:.6f} seconds") # Clarify timing
                print(f"  Number of steps: {len(cost_evo)-1}")

                # Generate static plot after algorithm execution and timing
                if len(edges) > 0: # If MST found
                    # Check if static plot should be skipped based on user's new criteria
                    if len(self.G.nodes()) > self.STATIC_PLOT_NODE_THRESHOLD and \
                       len(self.G.edges()) > self.STATIC_PLOT_EDGE_THRESHOLD:
                        print(f"Skipping static plot for {name} because graph has over {self.STATIC_PLOT_NODE_THRESHOLD} nodes AND over {self.STATIC_PLOT_EDGE_THRESHOLD} edges.")
                    else:
                        print(f"Generating static plot for {name}...")
                        self._plot_final_mst(edges, name, cost) # Call the static plot method
            except Exception as e:
                print(f"Error running {name}: {e}")
                traceback.print_exc()
                results[name] = {'failed': True, 'error': str(e)}
                # Continue to the next algorithm even if one fails
                continue

        print("\n--- Final Comparative Analysis Plots ---")
        self._display_comparison(results)
        self.plot_cost_evolution(all_alg_states)
        self.plot_edges_evolution(all_alg_states)
        self.plot_components_evolution(all_alg_states)
        self.plot_performance_comparison(results)

        # Call the new animation saving function after all comparisons
        self.save_mst_animation(mst_graphs_for_animation)

        plt.show(block=True) # Keep all generated plots open until user closes them
        return results

    def _verify_mst(self, mst_edges, calculated_cost):
        """
        Verifies the calculated MST against NetworkX's built-in MST implementation.
        Checks for cost, connectivity, and correct number of edges.
        """
        # Handle trivial graph cases (0 or 1 node)
        if not self.G.nodes(): # Original graph is truly empty (0 nodes)
            if calculated_cost == 0 and not mst_edges:
                print("✅ Graph is empty, MST is trivially correct (empty).")
            else:
                print("⚠️  Empty graph MST mismatch (cost or edges).")
            return
        elif len(self.G.nodes()) == 1: # Original graph has only 1 node
            if calculated_cost == 0 and not mst_edges:
                print("✅ Graph has 1 node, MST is trivially correct (empty).")
            else:
                print("⚠️  Single-node graph MST mismatch (cost or edges).")
            return


        # Create a NetworkX graph from the calculated MST edges
        mst_graph_from_alg = nx.Graph()
        mst_graph_from_alg.add_weighted_edges_from(mst_edges)

        # Get MST from NetworkX's built-in function for comparison
        nx_mst = nx.minimum_spanning_tree(self.G)
        nx_cost = sum(data['weight'] for u, v, data in nx_mst.edges(data=True))

        print("\nVerification:")
        print(f"- Your MST cost: {calculated_cost:.4f}")
        print(f"- NetworkX MST cost: {nx_cost:.4f}")

        # Compare costs using a small tolerance for floating-point precision
        if abs(calculated_cost - nx_cost) > 1e-6:
            print("⚠️  Cost mismatch detected!")
        else:
            print("✅ Cost matches NetworkX implementation")

        # Check connectivity of the generated MST and number of edges
        expected_nodes_in_mst = len(self.G.nodes())
        expected_edges_in_mst = expected_nodes_in_mst - 1

        # Check if the MST graph built by the algorithm is empty when it shouldn't be
        if mst_graph_from_alg.number_of_nodes() == 0 and expected_nodes_in_mst > 1:
            print("❌ MST graph built by algorithm is empty, but original graph has multiple nodes. MST failed to span the graph.")
        # Check if the MST graph has the expected number of nodes (should connect all original nodes)
        elif mst_graph_from_alg.number_of_nodes() != expected_nodes_in_mst:
            print(f"❌ MST graph has {mst_graph_from_alg.number_of_nodes()} nodes, expected {expected_nodes_in_mst} nodes (missing connectivity).")
        # Only check connectivity if the graph has nodes and is expected to be connected
        elif not nx.is_connected(mst_graph_from_alg):
            print("❌ MST is not connected!")
        else:
            print("✅ MST is properly connected")

        # Check if the number of edges in the generated MST is correct (N-1 for N nodes)
        # Only perform this check if the expected number of nodes is greater than 0
        if expected_nodes_in_mst > 0 and len(mst_graph_from_alg.edges()) != expected_edges_in_mst:
            print(f"⚠️  Edge count mismatch: {len(mst_graph_from_alg.edges())} edges vs expected {expected_edges_in_mst} for a connected MST.")
        else:
            print(f"✅ Correct number of edges ({expected_edges_in_mst})")

    def _display_comparison(self, results):
        """
        Displays a comprehensive comparison table of all MST algorithm results.
        """
        print("\n" + "="*80)
        print("ALGORITHM COMPARISON RESULTS")
        print("="*80)
        # Table header with formatting
        print(f"{'Algorithm':<25} {'Cost':<12} {'Time (s)':<12} {'Edges':<10} {'Steps':<10} {'Eff. (Cost/Time)':<18}")
        print("-"*80)

        # Filter out skipped or failed algorithms for display
        display_results = {name: data for name, data in results.items() if not data.get('skipped', False) and not data.get('failed', False)}

        if not display_results:
            print(f"{'No valid results available for comparison.':<78}")
            print("This might happen if all algorithms were skipped or failed.")
            print("="*80)
            return

        # Print data for each algorithm
        for name, data in display_results.items():
            # Calculate efficiency (cost per unit time)
            efficiency = data['cost'] / data['time'] if data['time'] > 0 else float('inf') # Avoid division by zero
            short_name = name.replace(' Algorithm', '') # Shorter name for display
            print(f"{short_name:<25} {data['cost']:<12.4f} {data['time']:<12.6f} "
                  f"{len(data['edges']):<10} {data['steps']:<10} {efficiency:<18.2f}")

        # Identify and display best performing algorithms
        # Ensure we only pick from algorithms that actually ran successfully
        successful_algorithms = {name: data for name, data in display_results.items() if data.get('time', 0) > 0 and data.get('cost', 0) > 0}

        if successful_algorithms:
            fastest = min(successful_algorithms.keys(), key=lambda x: successful_algorithms[x]['time'])
            cheapest = min(successful_algorithms.keys(), key=lambda x: successful_algorithms[x]['cost'])
            # Most efficient is based on Cost/Time (lower ratio is better)
            most_eff = min(successful_algorithms.keys(),
                           key=lambda x: successful_algorithms[x]['cost']/successful_algorithms[x]['time'] if successful_algorithms[x]['time'] > 0 else float('inf'))

            print("-"*80)
            print(f"🏆 Fastest: {fastest.replace(' Algorithm', '')}")
            print(f"🏆 Lowest Cost: {cheapest.replace(' Algorithm', '')}")
            print(f"🏆 Most Efficient (Cost/Time): {most_eff.replace(' Algorithm', '')}")
        else:
            print("\nNo successful algorithm runs to determine best performance.")
        print("="*80)

    def plot_performance_comparison(self, results):
        """
        Generates and displays bar charts comparing algorithm execution times and final MST costs.
        """
        # Filter out algorithms with no valid results (e.g., skipped or failed)
        plot_results = {name: data for name, data in results.items() if data.get('time', 0) > 0 and data.get('cost', 0) > 0 and not data.get('skipped', False) and not data.get('failed', False)}

        if not plot_results:
            print("No valid results to plot for performance comparison.")
            return

        algo_names = [name.replace(' Algorithm', '') for name in plot_results.keys()]
        times = [plot_results[name]['time'] for name in plot_results.keys()]
        costs = [plot_results[name]['cost'] for name in plot_results.keys()]

        # Create a figure with two subplots (one for time, one for cost)
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7), layout='constrained')
        fig.suptitle('MST Algorithm Performance Comparison', fontsize=18, fontweight='bold', color='#333333') # Original title color

        # --- Plot Execution Time (Left Subplot) ---
        ax1.set_facecolor('#f5f5f5') # Original background
        ax1.tick_params(axis='x', rotation=30, colors='black') # Original colors
        ax1.tick_params(axis='y', colors='black') # Original colors
        ax1.yaxis.label.set_color('black') # Original colors
        ax1.xaxis.label.set_color('black') # Original colors
        ax1.title.set_color('#555555') # Original colors
        ax1.grid(axis='y', linestyle='--', alpha=0.7, color='#cccccc') # Original grid color

        bars_time = ax1.bar(algo_names, times, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']) # Original colors
        ax1.set_title('Execution Time', fontsize=14)
        ax1.set_ylabel('Time (seconds)', fontsize=12)
        # Add value labels on top of each bar
        for bar in bars_time:
            yval = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2, yval + (max(times)*0.01), f'{yval:.4f}s', ha='center', va='bottom', fontsize=8, color='black') # Original text color

        # --- Plot MST Cost (Right Subplot) ---
        ax2.set_facecolor('#f5f5f5') # Original background
        ax2.tick_params(axis='x', rotation=30, colors='black') # Original colors
        ax2.tick_params(axis='y', colors='black') # Original colors
        ax2.yaxis.label.set_color('black') # Original colors
        ax2.xaxis.label.set_color('black') # Original colors
        ax2.title.set_color('#555555') # Original colors
        ax2.grid(True, color='#cccccc', linestyle='--', alpha=0.7) # Original grid color

        bars_cost = ax2.bar(algo_names, costs, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']) # Original colors
        ax2.set_title('Final MST Cost', fontsize=14)
        ax2.set_ylabel('Cost', fontsize=12)
        # Add value labels on top of each bar
        for bar in bars_cost:
            yval = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2, yval + (max(costs)*0.01), f'{yval:.2f}', ha='center', va='bottom', fontsize=8, color='black') # Original text color

        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent suptitle overlap

    def plot_cost_evolution(self, all_alg_states):
        """
        Generates a dedicated line plot showing the evolution of MST cost for all algorithms.
        """
        # Filter out algorithms with no valid cost evolution data
        plot_states = {name: data for name, data in all_alg_states.items() if 'cost_evolution' in data and len(data['cost_evolution']) > 1 and not data.get('failed', False) and not data.get('skipped', False)}

        if not plot_states:
            print("No valid cost evolution data to plot.")
            return

        fig = plt.figure(figsize=(12, 7), layout='constrained')
        ax = fig.add_subplot(111)

        # Revert to original style for this plot
        fig.patch.set_facecolor('white')
        ax.set_facecolor('#f9f9f9')
        ax.tick_params(axis='x', colors='black')
        ax.tick_params(axis='y', colors='black')
        ax.yaxis.label.set_color('black')
        ax.xaxis.label.set_color('black')
        ax.title.set_color('#333333')
        ax.grid(True, linestyle='--', alpha=0.7, color='#cccccc')

        fig.suptitle('Comparative MST Cost Evolution', fontsize=18, fontweight='bold', color='#333333')
        ax.set_xlabel('Steps', fontsize=12)
        ax.set_ylabel('Accumulated MST Cost', fontsize=12)

        colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2'] # Original colors
        markers = ['o', 's', '^', 'D', 'v', 'p', '*']
        linestyles = ['-', '--', ':', '-.', (0, (3, 5, 1, 5, 1, 5)), (0, (1, 1)), (0, (5, 1))]

        for i, (name, data) in enumerate(plot_states.items()): # Use filtered states
            ax.plot(range(len(data['cost_evolution'])), data['cost_evolution'],
                    label=name.replace(' Algorithm', ''),
                    color=colors[i % len(colors)],
                    marker=markers[i % len(markers)],
                    markersize=4, linewidth=2, alpha=0.8,
                    linestyle=linestyles[i % len(linestyles)])

        ax.legend(fontsize=10, loc='upper left', facecolor='white', edgecolor='black', labelcolor='black') # Legend for light bg
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    def plot_edges_evolution(self, all_alg_states):
        """
        Generates a dedicated line plot showing the evolution of edges in MST for all algorithms.
        """
        # Filter out algorithms with no valid edge evolution data
        plot_states = {name: data for name, data in all_alg_states.items() if 'edges_count_evolution' in data and len(data['edges_count_evolution']) > 1 and not data.get('failed', False) and not data.get('skipped', False)}

        if not plot_states:
            print("No valid edges evolution data to plot.")
            return

        fig = plt.figure(figsize=(12, 7), layout='constrained')
        ax = fig.add_subplot(111)

        # Revert to original style for this plot
        fig.patch.set_facecolor('white')
        ax.set_facecolor('#f9f9f9')
        ax.tick_params(axis='x', colors='black')
        ax.tick_params(axis='y', colors='black')
        ax.yaxis.label.set_color('black')
        ax.xaxis.label.set_color('black')
        ax.title.set_color('#333333')
        ax.grid(True, linestyle='--', alpha=0.7, color='#cccccc')

        fig.suptitle('Comparative Edges in MST Evolution', fontsize=18, fontweight='bold', color='#333333')
        ax.set_xlabel('Steps', fontsize=12)
        ax.set_ylabel('Number of Edges in MST', fontsize=12)

        colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2'] # Original colors
        markers = ['o', 's', '^', 'D', 'v', 'p', '*']
        linestyles = ['-', '--', ':', '-.', (0, (3, 5, 1, 5, 1, 5)), (0, (1, 1)), (0, (5, 1))]

        for i, (name, data) in enumerate(plot_states.items()): # Use filtered states
            ax.plot(range(len(data['edges_count_evolution'])), data['edges_count_evolution'],
                    label=name.replace(' Algorithm', ''),
                    color=colors[i % len(colors)],
                    marker=markers[i % len(markers)],
                    markersize=4, linewidth=2, alpha=0.8,
                    linestyle=linestyles[i % len(linestyles)])

        ax.legend(fontsize=10, loc='upper left', facecolor='white', edgecolor='black', labelcolor='black')
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    def plot_components_evolution(self, all_alg_states):
        """
        Generates a dedicated line plot showing the evolution of connected components for all algorithms.
        """
        # Filter out algorithms with no valid components evolution data
        plot_states = {name: data for name, data in all_alg_states.items() if 'components_count_evolution' in data and len(data['components_count_evolution']) > 1 and not data.get('failed', False) and not data.get('skipped', False)}

        if not plot_states:
            print("No valid components evolution data to plot.")
            return

        fig = plt.figure(figsize=(12, 7), layout='constrained')
        ax = fig.add_subplot(111)

        # Revert to original style for this plot
        fig.patch.set_facecolor('white')
        ax.set_facecolor('#f9f9f9')
        ax.tick_params(axis='x', colors='black')
        ax.tick_params(axis='y', colors='black')
        ax.yaxis.label.set_color('black')
        ax.xaxis.label.set_color('black')
        ax.title.set_color('#333333')
        ax.grid(True, linestyle='--', alpha=0.7, color='#cccccc')

        fig.suptitle('Comparative Connected Components Evolution', fontsize=18, fontweight='bold', color='#333333')
        ax.set_xlabel('Steps', fontsize=12)
        ax.set_ylabel('Number of Connected Components', fontsize=12)

        colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2'] # Original colors
        markers = ['o', 's', '^', 'D', 'v', 'p', '*']
        linestyles = ['-', '--', ':', '-.', (0, (3, 5, 1, 5, 1, 5)), (0, (1, 1)), (0, (5, 1))]

        for i, (name, data) in enumerate(plot_states.items()): # Use filtered states
            ax.plot(range(len(data['components_count_evolution'])), data['components_count_evolution'],
                    label=name.replace(' Algorithm', ''),
                    color=colors[i % len(colors)],
                    marker=markers[i % len(markers)],
                    markersize=4, linewidth=2, alpha=0.8,
                    linestyle=linestyles[i % len(linestyles)])

        ax.legend(fontsize=10, loc='upper right', facecolor='white', edgecolor='black', labelcolor='black')
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    def plot_degree_distribution(self):
        """
        Generates a bar chart showing the degree distribution of the loaded graph.
        """
        if not self.G or len(self.G.nodes()) == 0:
            print("Cannot plot degree distribution for an empty graph.")
            return

        degrees = [d for n, d in self.G.degree()]
        max_degree = max(degrees)

        fig = plt.figure(figsize=(10, 6), layout='constrained')
        ax = fig.add_subplot(111)

        # Revert to original style for this plot
        fig.patch.set_facecolor('white')
        ax.set_facecolor('#f9f9f9')
        ax.tick_params(axis='x', colors='black')
        ax.tick_params(axis='y', colors='black')
        ax.yaxis.label.set_color('black')
        ax.xaxis.label.set_color('black')
        ax.title.set_color('#333333')
        ax.grid(True, alpha=0.3, color='#cccccc')

        ax.set_ylabel('Number of Nodes', fontsize=12)

        # Binning for cleaner visualization
        bins = np.arange(0, max_degree + 2) - 0.5
        ax.hist(degrees, bins=bins, edgecolor='black', color='darkorange', alpha=0.8) # Original bar color

        ax.set_xlabel('Degree (number of connections)', fontsize=12)
        ax.set_title('Degree Distribution\nTotal Nodes: {}'.format(len(self.G.nodes())), fontsize=14)

        plt.tight_layout()
        plt.show(block=False) # Keep it open for later final show


class DisjointSetUnion:
    """
    Optimized Disjoint Set Union (DSU) data structure with path compression and union by rank.
    Also tracks the number of disjoint components.
    """
    def __init__(self, n):
        # Parent array: parent[i] stores the parent of element i. Initially, each element is its own parent.
        self.parent = list(range(n))
        # Rank array: rank[i] stores the rank (approximate height) of the tree rooted at i.
        self.rank = [0] * n
        # Number of connected components. Initially, N components for N elements.
        self.num_components = n

    def find(self, x):
        """
        Finds the representative (root) of the set containing element x.
        Applies path compression to flatten the tree.
        """
        if self.parent[x] != x: # If x is not the root of its set
            self.parent[x] = self.find(self.parent[x]) # Recursively find root and compress path
        return self.parent[x] # Return the root

    def union(self, x, y):
        """
        Unites the sets containing elements x and y.
        Uses union by rank heuristic to keep trees balanced.
        Returns True if a union operation successfully merged two distinct sets, False otherwise (x and y were already in the same set).
        Decrements num_components if a successful union occurs.
        """
        px, py = self.find(x), self.find(y) # Find roots of sets containing x and y

        if px == py:
            return False # x and y are already in the same set, no union needed

        # Union by rank heuristic: Attach the smaller rank tree under the root of the larger rank tree
        if self.rank[px] < self.rank[py]:
            px, py = py, px # Swap so px is the root of the "taller" or equal rank tree

        self.parent[py] = px # Make px the parent of py (merge py's set into px's set)

        if self.rank[px] == self.rank[py]:
            self.rank[px] += 1 # If ranks were equal, the new tree's rank increases

        self.num_components -= 1 # Successfully merged two distinct components into one
        return True

# --- Main Execution Block ---
if __name__ == "__main__":
    try:
        print("--- Running algorithms for netscience.csv (379 nodes, 914 edges) ---")
        print("--- This graph is small enough for static plotting and animation. ---")
        visualizer1 = MSTVisualizer(graph_file='/content/netscience.csv')
        # Animation is now enabled for netscience.csv
        results1 = visualizer1.compare_all_algorithms()

        # --- Adding a new large graph to demonstrate skipping static plot ---
        print("\n" + "="*80)
        print("--- Running algorithms for a LARGE SYNTHETIC GRAPH (designed to skip graph plots and animation) ---")
        print("--- This graph will be large enough to skip static MST plots and animation. ---")
        # Create a large graph that exceeds the static plot thresholds
        # Example: Erdos-Renyi graph (n nodes, p probability of edge)
        # N=6000, p=0.009 -> approx 162,000 edges. Both > 5000 nodes and > 14000 edges.
        num_nodes_large = 6000
        prob_edge_large = 0.009 # This will give a dense enough graph to exceed edge threshold
        print(f"Generating synthetic graph with {num_nodes_large} nodes and target edge density (p={prob_edge_large})...")
        large_graph = nx.erdos_renyi_graph(num_nodes_large, prob_edge_large, seed=42)

        # Ensure it's connected (Erdos-Renyi might not be for certain p)
        if not nx.is_connected(large_graph):
            print("Synthetic graph not connected, using largest connected component.")
            largest_cc = max(nx.connected_components(large_graph), key=len)
            large_graph = large_graph.subgraph(largest_cc).copy()

        # Assign random weights to the large graph edges
        for u, v in large_graph.edges():
            large_graph[u][v]['weight'] = random.uniform(0.1, 10.0)

        visualizer2 = MSTVisualizer(initial_graph=large_graph)

        # Animation and static plots will be skipped for this large graph automatically.
        results2 = visualizer2.compare_all_algorithms()

        print("\n--- All graph processing complete. Check plots and for generated GIF files. ---")

    except FileNotFoundError:
        print("\nError: The specified graph file was not found.")
        print("Please ensure your graph file (e.g., '/content/netscience.csv') is in the same directory as the script,")
        print("or provide its full path (e.g., '/content/netscience.csv').")
    except ValueError as ve:
        print(f"\nError: {ve}")
        print("The loaded graph is empty or has structural issues. Please check the content and format of your graph file.")
    except Exception as e:
        print(f"\nAn unexpected error occurred during execution: {e}")
        print("This might be due to issues with the graph data, or library installations.")
        print("Please ensure:")
        print("1. Your graph file (CSV or edgelist) is correctly formatted with valid nodes and weights (if applicable).")
        print("2. NetworkX, NumPy, Matplotlib, and Pandas libraries are all correctly installed and up to date.")
        print("3. The graph is sufficiently connected for MST algorithms to run.")
        print("\nDetailed traceback for debugging:")
        traceback.print_exc() # Print full traceback to help identify the exact issue
```
